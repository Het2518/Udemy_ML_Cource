```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
df = pd.read_csv("advertising.csv")
```

```python
auto = pd.read_csv('advertising.csv')
```

```python
auto.head()
```

```python
sns.distplot(auto['Daily Internet Usage'])
plt.show()
```

```python
sns.displot(x="Daily Internet Usage" , data=auto)
plt.show()
```

```python
sns.jointplot(x='Daily Internet Usage', y='Age', data=auto)
plt.show()
```

```python
#kind : { "scatter" | "reg" | "resid" | "kde" | "hex" }
sns.jointplot(x='Daily Internet Usage', y='Age', data=auto, kind='reg')
plt.show()
```

```python
sns.pairplot(auto[['Daily Time Spent on Site', 'Age', 'Area Income']])
```

```python
sns.stripplot(x="Age", y="Area Income", data=auto, jitter=True, hue="Clicked on Ad", palette="Set2")
```

```python
sns.stripplot(x="Age", y="Area Income", data=auto, hue="Clicked on Ad", palette="Set2")
```
